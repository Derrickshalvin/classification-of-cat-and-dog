```python
from keras.models import Sequential
from keras.layers import Conv2D,MaxPooling2D, Flatten, Dense, Dropout, BatchNormalization
from tensorflow.keras.regularizers import l2
```

```python
classification= Sequential()
```

```python
classification.add(Conv2D(32, 5, 5, input_shape = (256, 256, 3), activation = 'relu',kernel_regularizer=l2(l2=0.03)))
classification.add(BatchNormalization())
classification.add(MaxPooling2D(pool_size=(2,2), strides=(2,2)))
classification.add(Dropout(0.4))
```

```python
classification.add(Conv2D(64, 5, 5, activation='relu',kernel_regularizer=l2(l2=0.03)))
classification.add(BatchNormalization())
classification.add(MaxPooling2D(pool_size=(2,2), strides=(2,2)))
classification.add(Dropout(0.4))
```

```python
classification.add(Flatten())
```

```python
classification.add(Dense(32, activation='relu',kernel_regularizer=l2(l2=0.03)))
classification.add(Dropout(0.4))
classification.add(Dense(1, activation='sigmoid'))
```

```python
classification.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
```

```python
from keras.preprocessing.image import ImageDataGenerator
train_datagen = ImageDataGenerator(rescale = 1./255,horizontal_flip=True, shear_range=0.3, zoom_range=0.3)
test_datagen=ImageDataGenerator(rescale= 1./255)
```

```python
training_data= train_datagen.flow_from_directory('/content/gdrive/MyDrive/data/train',
                                                target_size=(256,256),
                                                batch_size=16,
                                                class_mode='binary')
```

```python
test_data= test_datagen.flow_from_directory('/content/gdrive/MyDrive/data/test',
                                                target_size=(256,256),
                                                batch_size=16,
                                                class_mode='binary')
```

```python
#! pip install livelossplot
```

```python
from livelossplot import PlotLossesKerasTF
```

```python
classification.fit(training_data, epochs=300,validation_data=test_data,callbacks=[PlotLossesKerasTF()])
```

```python
classification.save('img_class.h5')
```
